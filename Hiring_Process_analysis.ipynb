## Cleaning applicants

```python
import pandas as pd
applicants=pd.read_excel("Applicants.xlsx")
applicants.head()
```

```python
applicants.tail()
```

```python
applicants.info()
```

```python
# Detecting duplicates
duplicates = applicants[applicants.duplicated()]
duplicates
```

```python
applicants.nunique()
```

```python
# change data type of dates

applicants['Technical interview date'] = pd.to_datetime(applicants['Technical interview date'])
applicants['Board interview date'] = pd.to_datetime(applicants['Board interview date'])
applicants['Created'] = pd.to_datetime(applicants['Created'])
applicants['Modified'] = pd.to_datetime(applicants['Modified'])
```

```python
applicants.info()
```

```python
# Check for duplicate rows
duplicate_rows = applicants[applicants.duplicated()]

# Display duplicate rows (if any)
if not duplicate_rows.empty:
    print("Duplicate Rows:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

```python
applicants.isnull().sum()
```

```python
applicants['Technical interview date'] = applicants['Technical interview date'].fillna('1900-01-01')
applicants['Board interview date'] = applicants['Board interview date'].fillna('1900-01-01')
applicants['Created'] = applicants['Created'].fillna('1900-01-01')
applicants['Modified'] = applicants['Modified'].fillna('1900-01-01')
```

```python
applicants.to_excel('cleaned_applicants_final.xlsx', index=False)
```
